# Ćwiczenie 4a - Softmax, implementacja ze wzoru

Najczęściej wykorzystywaną funkcją **do klasyfikacji wieloklasowej**, w ostatniej warstwie sieci, jest **Softmax**. Jej podstawowym zadaniem jest zamiana zakresu wartości z O-neuronów, do zakresu [0-1], tak, aby wyniki sumowały się do 1, a zatem mogły być interpretowane jako prawdopodobieństwo przynależności do klasy. Formalnie:

$$
softmax(x): \mathbb{R}^K \rightarrow [0, 1]^K
$$

Funkcja Softmax opisywana jest wzorem:

$$
softmax(x_i) = \frac{exp(x_i)}{\sum_{j=0}^K x_k}
\tag{7}
$$

Innymi słowy: dla każdego x-a, wylicza się $exp(x_i)$ i dzieli przez sumę wszystkich $\sum_j^K exp(x_j)$.

Ze względu na fakt, że softmax wykorzystuje wartości ekspotencjalne `exp`, w praktyce, przed przystąpieniem do kalkulacji, stosuje się **skalowanie x**, żeby uniknąć błędu przepełnienia podczas obliczeń - `exp(bardzo-duża-liczba)`  może nie zmieścić się w pamieci.
Mając na uwadze to ograniczenie, softmax **w praktyce** liczony jest w sposób następujący:

$$
\begin{align*}
X_{norm} &= X - max(X)  \\ \\
softmax(x_{norm, i}) &= \frac{exp(x_{norm, i})}{\sum_{j=0}^K exp(x_{norm, j})}
\tag{7a}
\end{align*}
$$

# Ćwiczenie 4b - pochodna Softmax dla różnych funkcji kosztu

Pochodna funkcji Softmax rzadko jest używana *„sama dla siebie”*. Najczęściej łączy się ją z konkretną funkcją kosztu. Wynika to z faktu, że po odpowiednich przekształceniach, łatwo jest uzyskać uproszczoną postać pochodnej, skróconą ze względu np. na koszt *Cross Entropy*.

W tej sekcji, wyprowadzimy pochodną Softmax, w oderwaniu od konkretnej funkcji kosztu. Uzyskany wzór będzie nam przydatny w kolejnych sekcjach, gdy będziemy analizować konkretne funkcje kosztu.

Softmax przyjmuje postać, opisaną równaniem [7b] powyżej.
Do obliczenia pochodnej, wykorzystamy regułę **POCHODNEJ ILORAZU FUNKCJI**

$$
\frac{f(x)}{g(x)} = \frac{f'(x) g(x) - f(x)g'(x)}{g^2(x)}
$$

Gdzie $f'(x), g'(x)$ oznaczają skrótwo pochodne fukcji f i g po x.

W poniższych wyrpowadzeniach, zamiast pisać $X_{norm}$ jak we wzorze [7a], będziemy używać skrótu X.

Ponieważ Softmax zawiera w sobie iterację po poszczególnych indeksach, musimy rozpatrzeć dwa przypadki:

1. Gdy $i = j$:

$$
\begin{align*}
\frac{\delta softmax(x, i)}{\delta (x_i)} & \tag*{Bazowa postać funkcji Softmax} \\ \\
= \frac{\exp(x_i) \sum_{k=0}^K \exp(x,k) - \exp(x,i)\exp(x_i)}{\left(\sum_{k=0}^K \exp(x,k) \right)^2} & \tag*{Wykorzystanie reguły ilorazu, gdzie $ f(x) = \exp(x_i), g(x) = \sum_{k=0}^K \exp(x,k) $} \\ \\
= \frac{\exp(x_i) \sum_{k=0}^K \exp(x,k)}{\left(\sum_{k=0}^K \exp(x,k) \right)^2 } - \frac{\exp(x,i)\exp(x_i)}{\left(\sum_{k=0}^K \exp(x,k) \right)^2} & \tag*{Rozbicie ułamka na dwie części, żeby łatwiej było przetwarzać je osobno} \\  \\
= \frac{\exp(x, i)}{ \sum_{k=0}^K \exp(x,k)} - \left( \frac{\exp(x, i)}{ \sum_{k=0}^K \exp(x,k)} \right)^2 & \tag*{Skracamy oba ułamki i przekształcamy zbliżone fragmenty} \\ \\
= softmax(x, i) -(softmax(x, i))^2 & \tag*{Sprowadzamy do bazowej postaci - to jest po prostu funckja Softmax, z której wyszliśmy} \\ \\
= softmax(x, i)(1 - softmax(x, i))
\tag{8a}
\end{align*}
$$


2. Gdy $i \ne j$

$$
\begin{align*}
\frac{\delta softmax(x, i)}{\delta (x_j)} & \tag*{Bazowa postać funkcji Softmax} \\ \\
= \frac{0 - \exp(x,i) \exp(x,j)}{\left(\sum_{k=0}^K \exp(x,k) \right) ^2} & \tag*{Wykorzystanie reguły ilorazu, gdzie $ f(x) = \exp(x_i), g(x) = \sum_{k=0}^K \exp(x,k) $} \\ \\
= \frac{-\exp(x, i) \exp(x, j)}{\left(\sum_{k=0}^K \exp(x,k) \right) ^2} =  & \tag*{Uproszczenie, usunięcie zera} \\ \\
= \frac{-\exp(x, i)}{\left(\sum_{k=0}^K \exp(x,k) \right) ^2} \times \frac{\exp(x, j)}{\left(\sum_{k=0}^K \exp(x,k) \right)^2} & \tag*{Rozbicie na iloczyn dwóch ułamków} \\ \\
= - softmax(x, i) \times softmax(x, j)
\tag{8b}
\end{align*}
$$

Łączymy oba równania [8a] i [8b] i otrzymujemy końcową postać:

$$
\begin{align*}
\frac{\delta}{\delta x_i} softmax(x, k) = \begin{cases}
softmax(x, i)(1 - softmax(x, i)) & \text{Gdy $i = k$} \\
- softmax(x, i) \times softmax(x, k) & \text{Gdy $i \ne k$}
\end{cases}
\tag{8c}
\end{align*}
$$
